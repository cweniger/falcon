```python
import joblib
import numpy as np
import pylab as plt
```

```python
data = joblib.load("samples_posterior.joblib")
```

```python
class SampleCollection(list):
    def __init__(self, data):
        super().__init__(data)
        self.data = data

    def get(self, key, stack=True):
        samples = [v[key] for v in self.data]
        if stack:
            samples = np.stack(samples)
        return samples

samples = SampleCollection(data)
```

```python
z = samples.get('z')
```

```python
plt.scatter(z[:, 0], z[:, 1], s=1, alpha=0.5)
#plt.xlim([-.1, .1])
#plt.ylim([-.1, .1])
plt.axhline(0, color='red', lw=0.5)
plt.axvline(0, color='red', lw=0.5)
```

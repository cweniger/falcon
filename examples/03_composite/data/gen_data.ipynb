```python
from src.model import DrawFalconLetters, DrawSnake, SimulateData
import numpy as np
import pylab as plt
```

```python
sim1 = DrawFalconLetters(text="FAL")
sim3 = SimulateData(background_level = 10)

z1 = np.array([
    [-0.0, -0.5],
    [0.0, 0.0],
    [0.0, 0.5],
    ]).flatten()
img1 = sim1.simulate(z1)
x = sim3.simulate(img1)

plt.imshow(x)
plt.colorbar()
np.save("./data/x_obs_FAL.npy", x)
```
